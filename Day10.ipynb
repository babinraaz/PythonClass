```python
#matplot Library
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.linspace (0,20, 100)
plt.plot(x, np.sin(x),'-');
plt.plot(x, np.cos(x),'--')
plt.plot(x,np.log(x), '.');
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(2,1)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.log(x), linestyle='--')
```

```python
#use of ';' reomves the return value in graph
plt.plot(x, np.sin(x) , '--r');
plt.title( 'A nice sine Curve')
```
